```python
//Questão 1

package main

import "fmt"

func main() {
	var nota1, nota2, nota3 float64

	fmt.Print("nota 1: ")
	fmt.Scan(&nota1)
	fmt.Print("nota 2: ")
	fmt.Scan(&nota2)
	fmt.Print("nota 3: ")
	fmt.Scan(&nota3)

	media := (nota1 + nota2 + nota3) / 3

	if media >= 6 {
		fmt.Println("MEDIA =", media)
		fmt.Println("Aprovado")
	} else {
		fmt.Println("MEDIA =", media)
		fmt.Println("Reprovado")
	}
}
```

```python
//Questão 2

package main

import "fmt"

func main() {
	var nc int
	fmt.Print("Número de casos de testes: ")
	fmt.Scan(&nc)

	nct := 0
	for nc > nct {
		var p, pp, pg, pa, pc float64

		fmt.Println("Número de pessoas e as percentagens de Popular, Geral, Arquibancada e Cadeiras: ")
		fmt.Scan(&p)
		fmt.Scan(&pp)
		fmt.Scan(&pg)
		fmt.Scan(&pa)
		fmt.Scan(&pc)

		vt := ((p * pp / 100 * 1) + (p * pg / 100 * 5) + (p * pa / 100 * 10) + (p * pc / 100 * 20))
		fmt.Println("A RENDA DO JOGO N.", nct+1, " E =", vt)

		nct++
	}
}
```

```python
//Questão 5

package main

import "fmt"

func main() {
	var cc int
	var ca float64
	var tc string
	var vc float64

	fmt.Println("Digite o número da conta do cliente:")
	fmt.Scanln(&cc)

	fmt.Println("Digite o consumo de água em metros cúbicos:")
	fmt.Scanln(&ca)

	fmt.Println("Digite o tipo de consumidor (R - Residencial, C - Comercial, I - Industrial):")
	fmt.Scanln(&tc)

	if tc == "R" {
		vc = 5 + 0.05*ca
	} else if tc == "C" {
		if ca <= 80 {
			vc = 500
		} else {
			vc = 500 + 0.25*(ca-80)
		}
	} else if tc == "I" {
		if ca <= 100 {
			vc = 800
		} else {
			vc = 800 + 0.04*(ca-100)
		}
	}

	fmt.Println("CONTA =", cc)
	fmt.Println("VALOR DA CONTA =", vc)
}
```

```python
//Questão 6

package main

import "fmt"

func main() {
	var n, ne int

	fmt.Println("Número de temperaturas em Fahrenheit a serem convertidas para Celsius:")
	fmt.Scanln(&n)

	ne = 0

	for n > ne {
		var tconvertida, TC float64

		fmt.Println("Temperatura em Fahrenheit:")
		fmt.Scanln(&tconvertida)

		TC = 5 * (tconvertida - 32) / 9

		fmt.Println(tconvertida, "FAHRENHEIT EQUIVALE A", TC, "celsius")

		ne = ne + 1
	}
}
```

```python
//Questão 8

package main

import "fmt"

func main() {
	var r, h float64

	fmt.Println("Raio:")
	fmt.Scanln(&r)

	fmt.Println("Altura:")
	fmt.Scanln(&h)

	pi := 3.14159

	AC := pi * (r * r)
	AL := 2 * pi * r * h
	AT := 2*AC + AL
	C := AT * 100

	fmt.Println("O VALOR DO CUSTO E =", C)
}
```

```python
//Questão 18

package main

import "fmt"

func main() {
	var V1, r, ne, Vn, soma int

	fmt.Println("Escreva o valor inicial:")
	fmt.Scanln(&V1)

	fmt.Println("Escreva a razão:")
	fmt.Scanln(&r)

	fmt.Println("Escreva o número n de elementos da progressão:")
	fmt.Scanln(&ne)

	Vn = V1 + ((ne - 1) * r)
	soma = (V1 + Vn) * ne / 2

	fmt.Print("A soma é:",soma)
}
```

```python
//Questão 19

package main

import (
	"fmt"
)

func main() {
	var n int

	fmt.Println("Digite um número:")
	fmt.Scanln(&n)

	if n <= 1 {
		fmt.Println("Número inválido!")
	} else {
		soma := 0.0
		k := 1

		for k <= n {
			soma += 1.0 / float64(k)
			k++
		}

		fmt.Printf("%.6f\n", soma)
	}
}
```
